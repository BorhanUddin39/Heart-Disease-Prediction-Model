# Reading Data

```python
import pandas as pd

# Column names
column_names = ['Age', 'Sex', 'Chest Pain', 'Rest BP', 'Cholesterol', 'Fasting Sugar', 'Rest ECG', 'Max HR', 'Ex Angina',
                'ST Depression', 'ST Slope', 'Fluoroscopy', 'Thalassemia', 'Heart Disease']

# Reading the Hungarian data file
hungarian_path = '../data/processed.hungarian.data'
hungarian_df = pd.read_csv(hungarian_path, delimiter=',', header=None)
hungarian_df.columns = column_names

print("Number of Rows in Data    : ", hungarian_df.shape[0])
print("Number of Columns in Data : ", hungarian_df.shape[1])
display(hungarian_df.head())
```

```python
# Reading the Switzerland data file
switzerland_path = '../data/processed.switzerland.data'
switzerland_df = pd.read_csv(switzerland_path, delimiter=',', header=None)
switzerland_df.columns = column_names

print("Number of Rows in Data    : ", switzerland_df.shape[0])
print("Number of Columns in Data : ", switzerland_df.shape[1])
display(switzerland_df.head())
```

```python
# Reading the Cleveland data file
cleveland_path = '../data/processed.cleveland.data'
cleveland_df = pd.read_csv(cleveland_path, delimiter=',', header=None)
cleveland_df.columns = column_names

print("Number of Rows in Data    : ", cleveland_df.shape[0])
print("Number of Columns in Data : ", cleveland_df.shape[1])
display(cleveland_df.head())
```

```python
# Combining all dataframes
df = pd.concat([hungarian_df, switzerland_df, cleveland_df], ignore_index=True)

print("Number of Rows in Data    : ", df.shape[0])
print("Number of Columns in Data : ", df.shape[1])
display(df.head())
```

```python
# Saving the final combined data
df.to_csv('../data/heart_disease_combined.csv', index=False)
```
